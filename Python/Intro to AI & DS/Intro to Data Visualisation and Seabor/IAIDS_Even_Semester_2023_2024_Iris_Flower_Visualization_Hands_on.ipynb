```python
# import libraries
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

The dataset we will use in this exercise is the Iris Flower Dataset by Fisher. You can read more about this dataset [here](https://en.wikipedia.org/wiki/Iris_flower_data_set). There are five attributes in this dataset:


1.   sepal length (cm)
2.   sepal width (cm)
3.   petal length (cm)
4.   petal width (cm)
5.   class



There are 3 classes in the dataset, each class representing the class of the Iris flower. In this case, 0 corresponds to Iris setosa, 1 corresponds to Iris versicolor, and 2 corresponds to Iris virginica.

image.png

```python
# load iris data
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['class'] = iris.target # 0, 1, 2 correspond to 'setosa', 'versicolor', 'virginica'
df
```

```python
# barchart for counting
sns.countplot(x=df['class'])
```

**Question 1**: Create a barchart to show the number of data for each class. The bars should be in dark red color.

```python
sns.countplot(x=df['class'], color='darkred')
```

**Question 2**: Create a horizontal barchart to show the number of data for each class.

```python
sns.countplot(y=df['class'], color='darkred')
```

```python
# histogram of sepal length (cm)
sns.histplot(data=df, x='sepal length (cm)')
```

**Question 3**: Create a histogram to show the distribution of `sepal width (cm)`.

```python
sns.histplot(data=df, x='sepal width (cm)')
```

```python
# histogram with bin width of 1
sns.histplot(data=df, x='sepal width (cm)', binwidth=1)
```

**Question 4**: Create a histogram to visualize the distribution of `sepal width (cm)`. The bin width you should use is 0.1.

```python
sns.histplot(data=df, x='sepal width (cm)', binwidth=0.1)
```

```python
# density plot of sepal length (cm)
sns.kdeplot(data=df, x='sepal length (cm)')
```

**Question 5**: Create a density plot (KDE Plot) to show the distribution of all continuous features. Use only 1 plot to visualize all the distributions.

```python
sns.kdeplot(data=df, x='sepal length (cm)')
sns.kdeplot(data=df, x='sepal width (cm)')
sns.kdeplot(data=df, x='petal length (cm)')
sns.kdeplot(data=df, x='petal width (cm)')
plt.xlabel("Iris Flower Dataset")
```

```python
# density plot of petal length (cm) colored by class label
sns.kdeplot(data=df, x='petal length (cm)', hue='class')
```

**Question 6**: Create a density plot (KDE Plot) to show the visualization of `sepal width (cm)` for each class label.

```python
sns.kdeplot(data=df, x='sepal width (cm)', hue='class')
```

```python
# scatter plot
sns.scatterplot(data=df, x='petal length (cm)', y='petal width (cm)');
```

**Question 7**: Create a scatterplot to show the correlation between `petal length (cm)` as the x-axis and `petal width (cm)` as the y-axis. Group the color of datapoints by class, datapoint with different class should be in different color.

```python
sns.scatterplot(data=df, x='petal length (cm)', y='petal width (cm)', hue='class')
```

Boxplot is a plot used to visualize the locality, spread, and skewness groups of numerical data. You can use a boxplot to identify outliers in your dataset. You can read more about boxplot [here](https://en.wikipedia.org/wiki/Box_plot).

```python
# box plot of petal length (cm)
sns.boxplot(x='class', y='petal length (cm)', data=df);
```

```python
# box plot of sepal length (cm)
sns.boxplot(x='class', y='sepal length (cm)', data=df);
```

**Question 8**: Create a boxplot for `sepal width (cm)` for each class.

```python
sns.boxplot(x='class', y='sepal width (cm)', data=df)
```

The next visualization we will try to learn is heatmap. A heatmap is used to visualize the magnitude of individual values within a dataset as color. You can read more [here](https://en.wikipedia.org/wiki/Heat_map).

```python
# heatmap for correlation
corr = df[[ft for ft in df.columns if ft != 'class']].corr()
sns.heatmap(corr, annot=True)
```

**Question 9**: Create a similar heatmap to show the correlation between `sepal length (cm)` and `petal length (cm)` only.

```python
list = ['sepal length (cm)', 'petal length (cm)']
corr = df[[ft for ft in list if ft != 'class']].corr()
sns.heatmap(corr, annot=True)
```
